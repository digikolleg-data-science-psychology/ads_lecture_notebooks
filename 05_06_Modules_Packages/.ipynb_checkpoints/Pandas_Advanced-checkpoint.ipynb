The `DataFrame` has been converted from the regular *wide* format to the *long* format.

From Wikipedia (https://de.wikipedia.org/wiki/Wide-Format_und_Long-Format; translated):

* *Wide format* is well suited for displaying cross-sectional data or time series data. In the wide format, several columns contain the measured values of the same variable at different times or measurement repetitions (here, for example, different saliva measurements), while a Subject corresponds to one row of the table.

* In the *long format* all values of the measurement-repeated variable are represented in the *same* column and the corresponding time point in a *separate* variable. Therefore, the data in *long format* are also called "stacked". Other variables are noted in their own column, but also use the values of the time variable. The time variable (or *index variable*), which can also specify the test repetition or the context of the measurement, is thus explicitly specified.

<b>
Many statistic and plotting libraries require the usage of the long format, where the time variable is specified. Additionally, sorting and filtering data in the long format can be much more convenient. Therefore, the transformation from wide to long format is essential.
</b>

<div class="alert alert-block alert-info">
    <b>Hint:</b> Pandas <code>DataFrames</code> in the <i>long format</i> have an index variable in the MultiIndex data type.
</div>

## Transformation from *Wide-* to *Long-Format*

### Manually

```python
data = pd.read_csv('data/cortisol_sample.csv')
data.head()
```

Set index 

```python
index_cols = ['subject', 'condition']
data.set_index(index_cols, inplace=True)
# kann auch direkt beim Laden der csv-Datei übergeben werden: pd.read_csv('', index_col=index_cols)

data.head()
```

#### *Stack* the values 

The *wide format* DataFrame can be transformed into a long format DataFrame with `DataFrame.stack()`:

```python
pd.DataFrame(data.stack())
```

You will notice that the column has no name (preferred some meaningful name, in this case `cortisol`). Furthermore, the new index level does not have a name as well (preferred e.g. `sample`). Additionally, the index level is of type `str` not of type `int`.

All this operations can be performed either before or after stacking the DataFrame.

#### Change columns from `str` to `int`

```python
data.columns = data.columns.astype(int)
```

#### Change columns from `Index` to `MultiIndex`

To transform the DataFrame to long format later, the columns have to be in the `MultiIndex` format. This enables to write an index level as time variable later.

In this case, we want to create a new index level, which is created from a single value for each index entry. From [0, 1, ..., 6] we want to create the following `MultiIndex`: [('cortisol', 0), ('cortisol', 1), ..., ('cortisol', 6)], where the second index level (name: 'sample') will be the new column in the long format.

A `MultiIndex` can be constructed manually in many different ways:
* `pd.MultiIndex.from_arrays()`
* `pd.MultiIndex.from_tuples()`
* `pd.MultiIndex.from_product()`
* ...

Further readings: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.MultiIndex.html

In this case `pd.MultiIndex.from_product()` can be used:

```python
# column index before:
old_index = data.columns
old_index
```

```python
# column index after:
new_index = pd.MultiIndex.from_product([['cortisol'], data.columns], names=["", "sample"])
new_index
```

#### Set MultiIndex as new column index

```python
data.columns = new_index
data.head()
```

#### Final *long format* DataFrame

```python
data_long = data.stack()
data_long
```

### Transformation with `DataFrame.melt()`

Using `DataFrame.melt()` is another way of transforming a DataFrame from wide to long format.

Documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.melt.html

```python
data = pd.read_csv('data/cortisol_sample.csv')
data.head()
```

```python
data_long = data.melt(id_vars=['condition', 'subject'], value_name='cortisol', var_name="sample")
data_long.head()
```

Set the index column and sort index

```python
data_long = data_long.set_index(['subject', 'condition', 'sample'])
data_long = data_long.sort_index()
```

```python
data_long
```

### Transformation with `pd.wide_to_long()`

`pd.wide_to_long()` is another way of converting a DataFrame to *long format*. This function is easier to use but less flexible than `DataFrame.melt()`. `wide_to_long` expects column in the format `A-suffix1`, `A-suffix2`, `...`.

```python
data = pd.read_csv('data/cortisol_sample.csv')
data.head()
```

```python
data.set_index(["subject", "condition"], inplace=True)
data.columns = ["cortisol_{}".format(column) for column in data.columns]
data = data.reset_index()
data.head()
```

```python
data_long = pd.wide_to_long(data, stubnames="cortisol", i=['subject', 'condition'], j="sample", sep='_')
data_long
```

### Additional examples

Load DataFrame in the *long format*:

```python
data = bp.saliva.io.load_saliva('data/cortisol_sample.csv', biomarker_type='cortisol')
data
```

Calculation of some cortisol standard parameters: 

```python
data_features = bp.saliva.standard_features(data)
data_features.head()
```

```python
data_features = data_features.reset_index()
data_features.head()
```

#### Transform the DataFrames to *long format* with `DataFrame.melt()`

```python
data_long = data_features.melt(
    id_vars=['condition', 'subject'], 
    value_vars=['cortisol_argmax', 'cortisol_kurt', 'cortisol_mean', 'cortisol_skew', 'cortisol_std'], 
    var_name="feature", 
    value_name="cortisol"
)
# alternativ: keine Angabe der 'value_vars', dann werden alle Spalten gestacked:
# df_melt = data_features.melt(id_vars=['condition', 'subject'])
data_long.head()
```

Set & sort index:

```python
data_long = data_long.set_index(['subject', 'condition', 'feature']).sort_index()
data_long
```

### Using `pd.wide_to_long`

* `stubnames` (start of each column name): `cortisol`
* `sep` (separates `stubname` from `suffix`): `_`
* `i` (identifier columns, `id_vars`): `subject`, `condition`
* `j`: Name of the new index column in the long format: `biomarker`
* `suffix`: suffix format; as *regular expression (regex)*: `\w+` 

```python
data_long = pd.wide_to_long(
    data_features, 
    stubnames="cortisol", 
    sep='_', 
    i=['subject', 'condition'], 
    j='feature', 
    suffix=r"\w+"
)
data_long
```

### `GroupBy()`

Group DataFrames with a single or multiple criteria:
* index or column level
* a function, applied on index or columns
* ...

Documentation: https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html

```python
data_features.set_index("subject", inplace=True)
```

#### Basics

Example: Group with the column 'condition'. The result is an Object of the type `pd.DataFrameGroupBy`:

```python
data_features.head()
```

```python
grp = data_features.groupby('condition')
type(grp)
```

The `groups` attribut returns a `dict` with the different groups and the indeces of the elements that belong to that group:

```python
grp.groups
```

Number of elements in each group:

```python
grp.size()
```

Iterate through the groups:

```python
for key, df in grp:
    print(key)
    display(df.head())
```

Alternative: The `list()` operator on the `Groupby` object returns a list of the groups, where each group is a `tuple` with key and DataFrame:

```python
list(grp)
```

Select a single group:

```python
grp.get_group('Intervention').head()
```

#### Reductions

All reductions can applied analougously to a DataFrame object:

```python
grp.mean()
```

```python
grp.agg(['mean', 'std', 'min', 'max', np.argmax])
```

#### Transform

`transform` returns an object, which has the same size as the grouped object. `transform` is applied column-wise and is therefore suited for normalizations (*z-Transformation*, *log-Transformation*, *min-max-Normalisierung*).

```python
df_z_trans = grp.transform(lambda x: (x - x.mean()) / x.std())
df_z_trans.head()
```

```python
df_min_max = grp.transform(lambda x: (x - x.min()) / (x.max() - x.min()))
df_min_max.head()
```

#### Filtering

Group the data with the position of the maximum

```python
grp = data_features.groupby('cortisol_argmax')
```

```python
grp.size()
```

Filter the groups, that hava at least 5 elements

```python
df_filt = grp.filter(lambda x: len(x) >= 5)
```

```python
df_filt['cortisol_argmax'].unique()
```

### Interpolate missing data

```python
data = pd.DataFrame({
    'a': [1, 2, np.nan, 4, 5, 6],
    'b': [np.nan, 8, 9, 10, 11, 12],
    'c': [13, 14, np.nan, np.nan, 17, 18],
    'd': [19, 20, 21, 22, 23, np.nan],
})
data
```

Impute missing data column wise with linear interpolation

```python
data.interpolate()
```

Impute missing data row wise with linear interpolation

```python
data.interpolate(axis=1)
```

Impute missing data the last valid value

```python
data.interpolate(method='pad')
```

Only one consecutive missing value should be interpolated

```python
data.interpolate(limit=1)
```

Interpolate in both directions, to impute missing values at the start

```python
data.interpolate(limit_direction='both')
```
